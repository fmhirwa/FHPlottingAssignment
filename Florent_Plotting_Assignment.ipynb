```python
# YOUR CODE HERE
# HISTOGRAM of Carbon dioxide emissions
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the histogram

Co2_emission = df['co2']
sns.set(style="whitegrid")

sns.histplot(Co2_emission, bins=20, color='red', edgecolor='black')
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.title('Distribution of Carbon Dioxide Emissions')
plt.show()

# Print lines
print(Co2_emission.head(10))

```

# Interpretation of Engine Size Histogram:
##### The histogram of engine sizes illustrates the distribution of engine displacements across the sampled vehicles. The x-axis represents the engine displacement, typically measured in liters, while the y-axis denotes the frequency of occurrences. The graph highlights that most vehicles possess engine sizes within a common range, reflecting a concentration around median engine displacements. Conversely, very small or very large engines are less frequent, indicating that they are less common in the vehicle population.

```python
# YOUR CODE HERE
# HEATMAP FOR COMBINED FUEL EFFICIENCY
Combfuel= df['comb']
sns.histplot(data=df, x='comb', color='tan', edgecolor='black', bins=20)
plt.xlabel('Combined Fuel Efficiency')
plt.ylabel('Frequency')
plt.title('Distribution of Combined Fuel Efficiency')
plt.show()

# PRINT TEN LINES
print(Combfuel.head(10))

```

The histogram displays the distribution of combined fuel efficiency among vehicles. The x-axis represents fuel efficiency, and the y-axis shows the frequency of vehicles.
```python
# YOUR CODE HERE
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a heatmap
numbercolumns = df.select_dtypes(include=['number'])
corr_matrix = numbercolumns.corr()

# Create a heatmap
plt.figure(figsize=(9, 6))
sns.heatmap(corr_matrix,  cmap='coolwarm', fmt=".2f") #bwr and coolwarm are the closest I found
plt.title('Correlation Heatmap of Fuel Economy Data')

# SET TO ITALIC?

plt.show()


```

# Interpretation of the Heatmap Data
The heatmap visualizes the correlation coefficients between various vehicle attributes. Key insights include:

**High Correlation:** Certain features like comb (combined fuel efficiency), city, UCity, highway, and UHighway show strong positive correlations with each other, which is expected as they all relate to fuel efficiency metrics.
**Negative Correlations**: There are negative correlations between ghgScore and co2, cylinders, and displ. This suggests that vehicles with higher greenhouse gas scores tend to have lower CO2 emissions and smaller, less powerful engines.
